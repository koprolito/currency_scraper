<h1>Currency graphs</h1>

If needed, you can visualize how the prices of your currencies vary in the
following graphs

<h2>Hands on!</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

def get_all_previous_prices() -> dict[str:list[float]]:
        
        df = pd.read_csv('prices.csv', sep="|")
        previous_prices: dict[str:float] = {}

        for index, row in df.iterrows():
            prices = []
            for i in range(len(row)):
                if i > 0:
                    prices.append(float(row.iloc[i]))
            previous_prices[row.iloc[0]] = prices
        return previous_prices

previous_prices = get_all_previous_prices()
dates = []
for date in previous_prices:
    dates.append(date[:date.find(',')])
```

Now you can use the graphs ;)
<br>
Here is an example of how these graphs are implemented

```python
plt.style.use('dark_background')
plot_object = plt.subplots(nrows=1, ncols=1, figsize=(14, 6))
fig, ax_currency = plot_object
ax_currency.set_title('Dolar')
ax_currency.clear()

price_over_time = []
for date,prices in previous_prices.items():
    price_over_time.append(previous_prices[date][0])

ax_currency.plot(price_over_time,dates, c='green', label='Dolar')
ax_currency.legend()
```
